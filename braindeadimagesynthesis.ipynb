```python
import torch
from PIL import Image
from torchvision import transforms
from diffusers import StableDiffusionImageVariationPipeline
device = "cuda:0"
sd_pipe = StableDiffusionImageVariationPipeline.from_pretrained(
  "lambdalabs/sd-image-variations-diffusers",
  revision="v2.0",
  )
sd_pipe = sd_pipe.to(device)
```

```python
gen_img_dir = "/kaggle/working/Train"
!mkdir "/kaggle/working/Train"
!mkdir "/kaggle/working/Train/angry"
!mkdir "/kaggle/working/Train/sad"
!mkdir "/kaggle/working/Train/happy"
!mkdir "/kaggle/working/Train/neutral"
```

```python
from tqdm import tqdm
import os

train_dir = "/kaggle/input/brain-dead-emotion-detection/brain_dead_emotion_detection/Train"
for img_class in  [os.listdir(train_dir)[0]]:
    print(f"{img_class} is being generated...")
    for img in tqdm(os.listdir(os.path.join(train_dir, img_class)), desc=f"Image Class {img_class}"):
        im = Image.open(os.path.join(train_dir, img_class, img))
        tform = transforms.Compose([
            transforms.ToTensor(),
            transforms.Resize(
                (224, 224),
                interpolation=transforms.InterpolationMode.BICUBIC,
                antialias=False,
                ),
            transforms.Normalize(
              [0.48145466, 0.4578275, 0.40821073],
              [0.26862954, 0.26130258, 0.27577711]),
        ])
        inp = tform(im).to(device).unsqueeze(0)
        nsfw_content_detected = True
        while nsfw_content_detected == True:
            out = sd_pipe(inp, guidance_scale=3)
            nsfw_content_detected = out["nsfw_content_detected"][0]
        out["images"][0].resize((182, 182)).save(os.path.join(gen_img_dir, img_class, img))
```
